```python
# Check for missing values
print("Missing values per column:\n", credit_df.isnull().sum())

# Drop duplicates if any
credit_df.drop_duplicates(inplace=True)
print("After removing duplicates, shape:", credit_df.shape)
```

```python
# Countplot of fraud vs non-fraud
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=credit_df)
plt.title("Fraud vs Non-Fraud Transactions")
plt.show()

# Quantify class imbalance
print("Class distribution (normalized):")
print(credit_df['Class'].value_counts(normalize=True))
```

```python
# Transaction Amount distribution
plt.figure(figsize=(10,5))
sns.histplot(credit_df['Amount'], bins=50, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()

# Optional: log-transform to reduce skewness
credit_df['log_amount'] = np.log1p(credit_df['Amount'])
plt.figure(figsize=(10,5))
sns.histplot(credit_df['log_amount'], bins=50, kde=True)
plt.title("Log-Transformed Transaction Amount Distribution")
plt.show()
```

```python
# Select numeric columns
numeric_cols = credit_df.select_dtypes(include='number')

plt.figure(figsize=(14,10))
sns.heatmap(numeric_cols.corr(), annot=False, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Class', y='Amount', data=credit_df)
plt.title("Transaction Amount by Class (Fraud vs Non-Fraud)")
plt.show()
```

```python
# Time distribution of transactions
plt.figure(figsize=(12,5))
sns.histplot(credit_df['Time'], bins=100, kde=True)
plt.title("Transaction Time Distribution (Seconds from first transaction)")
plt.show()

# Fraud transactions over time
plt.figure(figsize=(12,5))
sns.histplot(credit_df[credit_df['Class']==1]['Time'], bins=100, color='red', label='Fraud', kde=True)
sns.histplot(credit_df[credit_df['Class']==0]['Time'], bins=100, color='blue', label='Non-Fraud', kde=True)
plt.title("Transaction Time Distribution by Class")
plt.legend()
plt.show()
```

```python
# Summary of key findings
print("Number of fraud transactions:", credit_df['Class'].sum())
print("Number of non-fraud transactions:", len(credit_df) - credit_df['Class'].sum())
print("Skewed features detected, log-transform applied to 'Amount'.")
print("Potential correlations between PCA features and fraud can be further analyzed.")
```
